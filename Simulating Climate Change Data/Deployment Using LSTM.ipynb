## ${\textbf{LSTM Forecasting under Climate Change scenarios}}$

```python
def createXY(dataset,n_past):
    dataX = []
    dataY = []
    for i in range(n_past, len(dataset)):
            dataX.append(dataset[i - n_past:i, 0:dataset.shape[1]])
            dataY.append(dataset[i,0])
    return np.array(dataX),np.array(dataY)
```

<a id="rcp4.5_1"></a>

>${\textbf{Representative Concentration Pathway :RCP 4.5 }}$

```python
RCP45 = pd.read_pickle('.././Pickles/RCP45/BNUESM.pkl')
RCP45["RG"]="RADES OUAFA"
RCP45["Zone"]=4
RCP45["GWL"]=0
RCP45["PZ"]="Bir El Kif"
column=['Time','GWL','RFm','RFt','RFs','RFy','PZ','RG','Zone','Year','Month','SPI','SPI_Category',]#Reindexing
RCP45=RCP45.reindex(column, axis='columns')
RCP45.index=RCP45.Time
del RCP45["Time"]
```

```python
RCP45
```

```python
lb_make = LabelEncoder()
RCP45["PZ"] = lb_make.fit_transform(RCP45["PZ"])
lb_make1 = LabelEncoder()
RCP45["RG"] = lb_make1.fit_transform(RCP45["RG"])
lb_make2 = LabelEncoder()
RCP45["SPI_Category"] = lb_make2.fit_transform(RCP45["SPI_Category"])
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
RCP45_scaled = scaler.fit_transform(RCP45)
```

```python
X,Y=createXY(RCP45_scaled,30)
```

```python
prediction=OldLSTM.predict(X)
prediction_copies_array = np.repeat(prediction,12, axis=-1)
pred=scaler.inverse_transform(np.reshape(prediction_copies_array,(len(prediction),12)))[:,0]

RCP45["PZ"] = lb_make.inverse_transform(RCP45["PZ"])
RCP45["RG"] = lb_make1.inverse_transform(RCP45["RG"])
RCP45["SPI_Category"] = lb_make2.inverse_transform(RCP45["SPI_Category"])
RCP45=RCP45[30:]
RCP45["GWL"]=pred*10
```
